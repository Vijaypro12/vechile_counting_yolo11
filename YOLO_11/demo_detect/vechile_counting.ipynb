```python
import cv2
from ultralytics import YOLO
from collections import defaultdict

model=YOLO('yolo11l.pt')

class_list=model.names
class_list
```

```python
cap=cv2.VideoCapture('test_video/4.mp4')

line_y_red = 430 

class_count= defaultdict(int)

crossed=set() 

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    #Run yolo tracking on the frame
    results= model.track(frame, persist=True, classes=[1,2,3,5,7])
    #print(results)

    #Ensure results ar(e not empty
    if results[0].boxes.data is not None:
        #get the detected boxes , their class indices
        boxes=results[0].boxes.xyxy.cpu()
        track_ids=results[0].boxes.id.int().cpu().tolist()
        class_indices=results[0].boxes.cls.int().cpu().tolist()
        confidence=results[0].boxes.conf.cpu()

        
        cv2.line(frame, (690, line_y_red), (1130, line_y_red), (0,0,255), 3)
      


        #loop through each detected object
        for box, track_id, class_indx, conf in zip(boxes, track_ids, class_indices, confidence):
            x1, y1, x2, y2 =map(int, box)
            cx=(x1+x2)//2
            cy=(y1+y2)//2
            class_name=class_list[class_indx]
            
            cv2.circle(frame, (cx, cy), 4, (0, 0, 0), -1)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 0,255), 2)
            
                            

            if cy > line_y_red and track_id not in crossed:
                crossed.add(track_id)
                class_count[class_name]+=1

            gap_bet=30
            for class_name, count in class_count.items():
                cv2.putText(frame, f"{class_name}: {count}", (50, gap_bet),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (290,56,34), 2)

                gap_bet+=30    


    cv2.imshow("YOLO Object Tracking & Counting Video", frame)  # Show frame
    
    # Press 'v' to quit
    if cv2.waitKey(1) & 0xFF == ord('v'):
        break

cap.release()
cv2.destroyAllWindows()
```
